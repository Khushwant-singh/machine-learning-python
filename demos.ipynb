## Missing values

```python
# find the number of missing values
data_df.isnull().sum()
```

```python
# missing values - percentage wise
(data_df.isnull().sum()/len(data_df))*100
```

```python
# removing the company feature
data_df.drop("company", inplace=True, axis=1)
```

```python
data_df.head()
```

```python
# unique values in children feature
data_df["children"].value_counts()
```

```python
# filling children feature with 0
data_df.children = data_df.children.fillna(0)
```

```python
# unique values in country feature
data_df["country"].value_counts()
```

```python
# dropping all rows with missing values in the country feature
data_df = data_df.dropna(subset=["country"])
```

```python
data_df
```

```python
# resetting the index
data_df = data_df.reset_index(drop=True)
data_df
```

```python
# agent feature
data_df["agent"].value_counts()
```

```python
# unique values in agent feature
data_df.agent.nunique()
```

```python
data_df.isnull().sum()
```

## Duplicates

```python
# total number of duplicates in the dataset
data_df.duplicated().sum()
```

```python
# extract the dataframe containing the duplicate values
data_df.loc[data_df.duplicated(), :]
```

```python
# dropping duplicate rows
data_df.drop_duplicates(inplace=True, keep="first")
```

```python
data_df.shape
```

## Datetime

```python
data_df["reservation_status_date"] = pd.to_datetime(data_df["reservation_status_date"])
```

```python
data_df.info()
```

```python
# get year, month and day
data_df["reservation_status_year"] = data_df["reservation_status_date"].dt.year
data_df["reservation_status_month"] = data_df["reservation_status_date"].dt.month
data_df["reservation_status_day"] = data_df["reservation_status_date"].dt.day
data_df.head()
```

```python
#week of year, day of week
data_df["res_week_of_year"] = data_df["reservation_status_date"].dt.week
data_df["res_day_of_week"] = data_df["reservation_status_date"].dt.dayofweek
data_df.head()
```

```python
#dropping the reservation status date feature
data_df.drop("reservation_status_date", inplace=True, axis=1)
```
