```python
df = df.drop(columns=["Cabin", "Ticket"])
```

```python
print(df.columns)
```

```python
embarked_mode = df["Embarked"].mode()[0]
print(embarked_mode)
```

```python
print(df[df["Embarked"].isna()].head())
df["Embarked"] = df["Embarked"].fillna(embarked_mode)
print(df[df["Embarked"].isna()].head())
```

```python
age_median = df["Age"].median()
print(age_median)
```

```python
df["Age"] = df["Age"].fillna(age_median)
```

```python
print(df.isna().sum())
```

```python
before_rows = len(df)
df = df[df["Fare"]>0]
after_rows = len(df)
print("Rows removed with non-positive Fare:", before_rows - after_rows)
```

```python
print(df.dtypes)
df["Sex"] = df["Sex"].astype("category")
df["Embarked"] = df["Embarked"].astype("category")
df["Pclass"] = df["Pclass"].astype("category")
df["Survived"] = df["Survived"].astype("int8")
print(df.dtypes)
```

```python
# Overall survival rate
```

```python
total = len(df)
survived = df["Survived"].sum()
print("Total passengers:", total)
print("Survived:", survived)
print("Survival rate: ", (survived / total))
```

```python
survival_by_sex = df.groupby("Sex")
```

```python
print(survival_by_sex)
```

```python
survival_by_sex = df.groupby("Sex").mean()
print(survival_by_sex)
```

```python
survival_by_sex = df.groupby("Sex")["Survived"].mean()
print(survival_by_sex)
```

```python
survival_by_class = df.groupby("Pclass")["Survived"].mean()
print(survival_by_class)
```

```python
avg_age_by_survival = df.groupby("Survived")["Age"].mean()
print(avg_age_by_survival)
```

```python
women_third = df[(df["Sex"]== "female") & (df["Pclass"] == 3)]
print("Count of women in third class:", len(women_third))
print("Survival rate: ", women_third["Survived"].mean())
```

```python
women_first = df[(df["Sex"]== "female") & (df["Pclass"] == 1)]
print("Count of women in first class:", len(women_first))
print("Survival rate: ", women_first["Survived"].mean())
```

```python
# loop through 
```

```python
genders = ["female", "male"]
classes = [1, 2, 3]

for gender in genders:
    for cls in classes:
        subset = df[(df["Sex"] == gender) & (df["Pclass"] == cls)]
        
        print(f"{gender.capitalize()} - Class {cls}")
        print("Count:", len(subset))
        print("Survival rate:", subset["Survived"].mean())
        print("-" * 30)
```

```python
## Option 2: Pandas-recommended way (groupby)
```

```python
summary = (
    df
    .groupby(["Sex", "Pclass"])
    .agg(
        Count=("Survived", "size"),
        Survival_rate=("Survived", "mean")
    )
)

print(summary)
```

```python
##âœ… Option 3: Loop but use query() (more readable)
```

```python
for gender in ["female", "male"]:
    for cls in [1, 2, 3]:
        subset = df.query("Sex == @gender and Pclass == @cls")
        print(f"{gender.capitalize()} - Class {cls}: "
              f"Count={len(subset)}, "
              f"Survival rate={subset['Survived'].mean():.2f}")
```
